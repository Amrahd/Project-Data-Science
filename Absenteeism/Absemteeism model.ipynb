# Pembentukan Model 

model yang di bentuk menggunakan regresi logistik

```python
import pandas as pd
import numpy as np 
```

```python
data_preprocessed = pd.read_csv('Absenteeism_preprocessed.csv')
```

```python
data_preprocessed.head(7)
```

```python
data_preprocessed['Absenteeism Time in Hours'].median()
```

```python
targets = np.where(data_preprocessed['Absenteeism Time in Hours'] > 
                   data_preprocessed['Absenteeism Time in Hours'].median(),1,0)
```

```python
data_preprocessed['Excessive Absenteeism'] = targets
```

```python
data_preprocessed.head(7)
```

```python
data_with_targets = data_preprocessed.drop(["Absenteeism Time in Hours"], axis= 1)
```

```python
data_with_targets
```

```python
unscaled_inputs = data_with_targets.iloc[:,:-1]
```

```python
unscaled_inputs
```

```python
y = data_with_targets.iloc[:,14:]
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler

absenteeism_scaler = StandardScaler()
```

```python
absenteeism_scaler.fit(unscaled_inputs)
```

```python
unscaled_inputs[['Transportation Expense', 'Distance to Work',
       'Age', 'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Month_Value','Day of the Week']] = absenteeism_scaler.fit_transform(unscaled_inputs[['Transportation Expense', 'Distance to Work',
       'Age', 'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Month_Value','Day of the Week']])
```

```python
scaled_inputs = unscaled_inputs.copy()
```

```python
scaled_inputs
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(scaled_inputs, y, train_size=0.2, random_state=20)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
reg = LogisticRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```
